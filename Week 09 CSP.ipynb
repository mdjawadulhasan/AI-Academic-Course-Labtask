```python
# # Define a new cgraph
# cgraph = {'WA': ['SA', 'NT'], 
#           'Q': ['SA', 'NT', 'NSW'], 
#           'T': [],
#           'V': ['SA', 'NSW'], 
#           'SA': ['WA', 'NT', 'Q', 'NSW','V'], 
#           'NT': ['Q', 'WA', 'SA'], 
#           'NSW': ['Q', 'V', 'SA']}

# Taking a Cgraph of my own area in Dhaka

cgraph = {'BS': ['RA', 'AF'], 
          'RA': ['BS', 'AF','HAT','B'], 
          'AF': ['BS','RA','BD'],
          'B': ['RA', 'AF','HAT','G'], 
          'HAT': ['RA', 'B', 'G'], 
          'G': ['HAT', 'B'], 
          'KAK': []}


```

```python
import random
def get_color(node, rv, cg):
    # CODE
    # check rv[node] is empty or not when no color availabe
    # return 'no color'
    if (len(rv[node])==0):
        return "no color"
        
    
    color = random.choice(rv[node])
    rv.pop(node)
    # update remaining values for neihbouring node
    adjList = cg[node]
    for adj in adjList:
        if adj in rv.keys():
            rv[adj].remove(color)
    #print(adjList)
    return color
```

## Highest Degree Node

```python
# write a method to get the highest degree node
def get_hdnode(cg, solved):
    mnode = ''
    degree = -1
    for k in cg.keys():
        if len(cg[k])  > degree and k not in solved.keys():
            #print(k)
            degree = len(cg[k])
            mnode = k
    return mnode
```

Color the nodes using Highest degree heuristic

```python
assigned = {} # store assignes variables with colors
# remainingValues = {'WA': ['R','G','B'], 
#                     'Q': ['R','G','B'],
#                     'T': ['R','G','B'],
#                     'V':  ['R','G','B'], 
#                     'NT': ['R','G','B'],
#                     'SA': ['R','G','B'],                    
#                     'NSW': ['R','G','B']}


remainingValues = {'BS': ['R','G','B'], 
                    'RA': ['R','G','B'],
                    'AF': ['R','G','B'],
                    'B':  ['R','G','B'], 
                    'HAT': ['R','G','B'],
                    'G': ['R','G','B'],                    
                    'KAK': ['R','G','B']}

for i in range(len(cgraph)):
    # assign color to the highest degree node
    hdNode = get_hdnode(cgraph, assigned)
    color = get_color(hdNode, remainingValues, cgraph)
    # CODE
    # check if 'no color' is returned
    # print failure
    
    if(color=="no color"):
        print("failure")
        break
    

    assigned[hdNode]= color
#     print(remainingValues)
    print(assigned)
```

## Minimum Remaining Value

```python
# define a method to select a node by MRV
# CODE
#implement the logic on remainingValues list

import sys
def get_MRVnode(rv):
    minnode = ''
    value = sys.maxsize
    for k in rv.keys():
        if len(rv[k]) < value:
            value = len(rv[k])
            minnode = k
    return minnode
```
