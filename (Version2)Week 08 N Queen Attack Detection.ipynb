# N Queen Attack Detection

```python
import random
N = 8
r = []
c = []


r.append(random.choice(range(1, N+1)))
c.append(random.choice(range(1, N+1)))
print('Q{}: {},{}'.format(1,r[-1],c[-1]))

i=2
while i < (N + 1):
    flag = 1
    temp_r=(random.choice(range(1, N+1)))
    temp_c=(random.choice(range(1, N+1)))

    # print(r)
    # print(c)
    # print("random Valu: r, c :", temp_r, temp_c)

    for j in range (len(r)):

        # print("Checking : ")
        # print("r[j], c[j] :",r[j],r[i])

        if ( (temp_r == r[j] ) and (temp_c == c[j]) ):
            flag=0
            # print("Duplicate Found")
            break

    if(flag==1):
        r.append(temp_r)
        c.append(temp_c)
        print('Q{}: {},{}'.format(i, r[-1], c[-1]))
        i += 1

print(r)
print(c)
```

```python
# row attack
for i in range(0,N):
    for j in range(i+1,N):
        if(r[i] == r[j]):
            print('row attack: row:{} column:{} and {}'.format(r[i],c[i],c[j]))      
```

```python
#column attack
for i in range(0,N):
    for j in range(i+1,N):
        if(c[i] == c[j]):
            print('Col attack: Col:{} row:{} and {}'.format(c[i],r[i],r[j])) 
```

```python
# diagonal atack

for m in range(0, N):
    for n in range(m + 1, N):
        i=r[m]
        j=c[m]
        k=r[n]
        l=c[n]
#         print(i,j,k,l)
        if(abs(j-l)==abs(k-i)):
            print("Attack Found")
            print("Row , Col :",i,j)
            print("Row , Col :",k,l)
            
            

        
```

```python
def Check_Attack(r,c):
    for i in range(0, N):
        for j in range(i + 1, N):
            if (r[i] == r[j]):
                return False

    for i in range(0, N):
        for j in range(i + 1, N):
            if (c[i] == c[j]):
                return False

    for m in range(0, N):
        for n in range(m + 1, N):
            i = r[m]
            j = c[m]
            k = r[n]
            l = c[n]
            if (abs(j - l) == abs(k - i)):
               return False
        

        
```

```python
b=(Check_Attack(r,c))
if(b==False):
    print("Attack Found")
else:
    print("No Attack Found")
```
